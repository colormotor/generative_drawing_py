## Sequences: lists and tuples



Last week we treated loops in a way that is closer to JavaScript (e.g. counting
from 0 to another number). But we also have seen that in Python we should
actually conceptualize a loop as an operation that **iterates** over sequences of
items, i.e. **it picks one item at a time from a sequence**.



### Lists



To dig deeper into this concept, it is a good idea to start from a very
important Python data structure: the `list`.

Lists in Python are similar to JavaScript&rsquo;s `Array` object, which you might have
seen already. Lists are like &ldquo;containers&rdquo; that allow you to store a sequence of
multiple values into a single variable. The word &ldquo;sequence&rdquo; is important here
because the items in a list appear in a specific order, so you have a first item,
a second item and so forth.

Let&rsquo;s create a simple list, by using the &ldquo;square bracket notation&rdquo;



````python
my_list = ['first', 'second', 'third', 'fourth']
print(my_list)
````

Here we create a list of items by placing the items within square brackets and
separating them with commas.

The list has a number of elements or size, which can be queried using the Python
built-in `len` function:



````python
len(my_list)
````

#### Accessing list elements



Single elements can be accessed using an &ldquo;index&rdquo; in the list, starting from
zero. For example



````python
my_list[0]
````

will give us the first element in the list.

Remember, **to access an element of a list we use the variable name followed by
square brackets containing the index**. Notice that, similarly to JavaScript and
similarly to the way in which we constructed loops, indices start from zero. As
a result the last element of a list can be accessed with the index
`len(my_list)-1`



````python
my_list[len(my_list)-1]
````

##### Extra info



Writing `len(my_list)` everytime you want to access the end of a list can be annoying. Python helps us save precious energy when typing, so we can simply use negative indices and access the list &ldquo;backwards&rdquo;. As an example, the last element can be also accessed with



````python
my_list[-1]
````

When you see a negative index within square brackets, imagine putting the length of the list before the negative sign.



#### Setting elements



You can use the index syntax to **set** elements of a list, for example



````python
my_list[0] = 100
print(my_list)
````

Will set the first element of a list to be `100`.



#### Iterating over lists



Once we have a list we can use a `for` loop to iterate over its elements.
As an example of how this can be useful, let&rsquo;s create a list of RGB colors and draw
a random circle for each color. We can represent each color with a list of three values, one for each RGB color component:



````python
create_canvas(300, 300)
background(0)
colors = [[255, 0, 0], [0, 255, 0], [0, 0, 255]]
for color in colors:
    fill(color)
    circle(random(0, width), random(0, height), 100)
show()
````

Note that above we have a list of lists, where each color is represented as a tripled of red, green ad blue values. The `fill` function also accepts a single list, or any type of other sequence of numbers that we will see (e.g. tuples and vectors).

Again, there is not a big conceptual difference in this approach of using `for` loops from the `range` syntax we used previously, since `range` spits out an ordered sequence of numbers. We can view what these numbers are by converting the range to a list, by using the type `list` as a function:



````python
list(range(10))
````

#### Lists methods and Python &ldquo;objects&rdquo;



Before we go on we need to touch on the concept of an &ldquo;object&rdquo; in programming.
What is an object? In Python, almost every variable you create, including lists,
is an &ldquo;object&rdquo;. It is a fancy way of saying a &ldquo;thing&rdquo; that has some properties
and can do certain actions. Objects have:

-   Attributes (information about them)
-   Methods (things they can do, functions specific to the object)

The way in which we access attributes and methods is the &ldquo;dot notation&rdquo;. E.g. if you had an object representing a virtual ball, something like

    ball.bounce()

will perform the &ldquo;bounce&rdquo; function specific to the instance of the object &ldquo;ball&rdquo;. The function &ldquo;bounce&rdquo; is often referred to as a method. The difference from a &ldquo;normal&rdquo; function is that a method depends on the properties of the object it is attached to.

For attributes, we have a similar notation without the parentheses, so

    c = ball.color

will access the value of the attribute `color` specific to the `ball` object and place it into a new variable called `c`.



##### Dynamically manipulating lists



With the &ldquo;dot notation&rdquo; in mind, we can dynamically create lists by first
creating an empty list:



````python
my_list = []
print(my_list)
````

And then add elements to the list dynamically using the `.append` method.



````python
for i in range(5):
    my_list.append(i)
print(my_list)
````

in this case we created a loop that adds an element to the list at each iteration.
If you try to run this code again, you will that the list will keep on growing.

We can also remove elements from the the list using the `.pop` method. With no arguments, `pop` removes the last element of the list and returns it. E.g.



````python
my_list.pop()
````

removes the last element and prints it out because a notebook cell always prints
out the result of the last instruction in the cell. Now the list is one element
shorter



````python
print(len(my_list))
````

We can also remove the first element (i.e. the zero&rsquo;th) by providing `0` as an argument, e.g.



````python
my_list.pop(0)
````

or any other element by index. Note that in most practical scenarios you will
probably need to remove the first or the last element of a list (no argument or
0). We will use this in the lab to create a drawing program that saves vector images instead of raster ones.



### Tuples



Another common way to describe sequences of items in Python is the `tuple`.
Tuples behave similarly to lists but **do not allow modification**. In technical
jargon we say that tuples are *immutable*.
To create a tuple you replace the square brackets with parentheses:



````python
my_tuple = ('first', 'second', 'third')
````

and we can access the elements or the length of the tuple similarly to lists. So



````python
print(my_tuple[0])
print(len(my_tuple))
````

However, now we are not able to modify an element of the tuple, so this will throw an error:



````python
my_tuple[0] = 100
````

Tuples can be initialized without parenthes, making them an intrinsic part of the Python syntax.
For example the above could have been:



````python
my_tuple = 'first', 'second', 'third'
````

This is particularly useful since we can make multiple assignments in one line using this method:



````python
r, g, b = 10, 20, 30
````

Where each element on the left of the `=` is being assigned the corresponding element on the right. This can allow you to quicly swap two values, e.g



````python
r, g = g, r
````

Where `g` gets assigned to `r` and `r` gets assigned to `g`. Doing this without this syntax would require you to use a temporary variable to store one of the two values you want to swap.

We can also &ldquo;distribute&rdquo; values from a tuple into multiple variables, e.g



````python
rgb = (255, 100, 0)
r, g, b = rgb
````

where we first create a tuple called `rgb` with three values, and then assign the values in order to three variables called `r`, `g` and `b`.

We can use this kind of syntax in loops to accesss the elements of a tuple in a sequence. As a practical example we can rewrite the random circle example as follows:



````python
create_canvas(300, 300)
background(0)
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
for r, g, b in colors:
    fill(r, g, b)
    circle(random(0, width), random(0, height), 100)
show()
````

## Linear interpolation, `lerp` and `linspace`



In earlier sessions, we have seen the `remap` function, which acts as a &ldquo;translator&rdquo; that
takes a value in one range and places it proportionally into another range. One
useful analogy that we did not see before is comparing `remap` to a horizontal slider
of a user interface. In that case the slider &ldquo;remaps&rdquo; a horizontal position
along the slider to the minimum and maximum value of the slider.

Internally this remapping procedure relies on a procedure called &ldquo;linear
interpolation&rdquo; (often referred to as &ldquo;lerp&rdquo;).



### Lerp



Linear interpolation is the process of finding points between two or more
points. Linear interpolation finds a position between a starting and an ending
point as if there were a line connecting them. This can be done by specifying
the relative &ldquo;percentage&rdquo; between the two points expressed as a value between 0
and 1 (with 1 = 100%), then linear interpolation is simply

```python
start + (end - start)*percentage
```

We can do this manually, or use the function `lerp(start, end, amt)` where `amt`
is the percentage. For example let&rsquo;s draw a circle on a position along a
horizontal line with a circle at a certain percentage along the line:



````python
create_canvas(400, 100)
background(0)
translate(width/2, height/2)
start = -100
end = 100
amt = 0.9
fill(0, 128, 255)
stroke(255)
line(start, 0, end, 0)
circle(lerp(start, end, amt), 0, 10)
show()
````

Try modifying the value of the `amt` variable. If the value of `amt` is less or
greater than 1 then the circle is not on the line segment anymore, which
technically is called &ldquo;extrapolation&rdquo;.

We actually have seen an example of using `lerp` (written manually) in the drawing example of Week 2. There we were find a line between the current &ldquo;brush&rdquo; position and the current mouse position and using linear interpolation to move a small amount along that line. More on that later in this document.



### Linspace



Now, let&rsquo;s see another function that does linear interpolation and will make creating grids much easier. This is where we start using the power of Python in a way that would not be as simple in P5js.

Let&rsquo;s say we do not want one value between a start and an end, but rather a sequence of equally spaced values. We did something like that to create grids, but we used the index `i` of a loop with `range`, which required us to use `remap` or other calculations to find points on the grid.

For example for to create a sequence of rectangles we did something like this:



````python
create_canvas(400, 100)
background(0)
fill(255, 0, 0)
n = 15
spacing = width / (n - 1)
for i in range(n):
    rect(i*spacing, 0, 10, height)
show()
````

where we had to calculate the horizontal position based on the index `i`. We can simplify this greatly by using the function `linspace(start, end, number_of_steps)`, which gives us a sequence of numbers equally spaced between `start` and `end` (the first two arguments). For example



````python
linspace(0, 1, 5)
````

gives us a sequence of numbers going from 0 to 1 in 5 equal steps. Each step is `1/(number_of_steps-1)`, which is equal to `0.25`.

Do not worry for now about the &ldquo;array&rdquo; that appears in the output. Just keep in mind that this is a sequence of numbers that we can iterate over using loops.

The earlier code now becomes:



````python
create_canvas(400, 100)
background(0)
fill(255, 0, 0)
for x in linspace(0, width, 15):
    rect(x, 0, 10, height)
show()
````

The `linspace` has an optional fourth parameter that allows to skip the last element if we set it to `False`:



````python
linspace(0, 1, 5, False)
````

We will use this in the labs to draw shapes that close smoothly.

By using `linspace`, drawing on a grid can become something like:



````python
create_canvas(400, 400)
margin = 30
for y in linspace(margin, height-margin, 10):
    for x in linspace(margin, width-margin, 10):
        circle(x, y, 5)
show()
````

### `arange`



Using `linspace` is convenient, but for creating a grid we could do better. To create rectangles in the example above we would have had to calculate the spacing for a given number of grid cells. We can avoid doing so with another function that gives us a sequence of evenly spaced numbers `arange(start, stop, step)`. It gives us a sequence of numbers between `start` and `stop` with equal spacing `step`. The main difference to keep in mind with `linspace` is that `arange` will not reach the value in `stop`, e.g.



````python
arange(0, 1, 0.25)
````

which does not contain the stop value, while this will



````python
arange(0, 1+0.01, 0.25)
````

So to create a grid of squares now we can do:



````python
create_canvas(400, 400)
tile_size = 20
for y in np.arange(0, width, tile_size):
    for x in np.arange(0, width, tile_size):
        square(x, y, tile_size)
show()
````

## Vectors



In the examples above we saw how to use linear interpolation to find a point along a horizontal line. We were referring to the values on the horizontal as &ldquo;points&rdquo;, but what if we wanted to find a point along a line that is not horizontal? If we think of coordinates we could do this with **two linear interpolations**, one for the X coordinate and one for the Y, e.g.



````python
create_canvas(400, 200)
background(0)
start_x = 200
start_y = 30
end_x = 300
end_y = 150
amt = 0.3
fill(0, 128, 255)
stroke(255)
arrow(start_x, start_y, end_x, end_y)
circle(lerp(start_x, end_x, amt),
       lerp(start_y, end_y, amt),
       10)
show()
````

Here we use `arrow` instead of `line` to indicate the direction of movement.
In practice, this arrow is called a **vector**. Think of it as an arrow in space (in our case 2d) that has a length and a direction. It encodes the **distance** and the direction of the displacement from its tail to its tip. The tip of the arrow pointing in the direction of the vector.

In 2d you can think of something like this:



In practice, we have been working with vectors all along. For example, take the position of a point on the canvas with coordinates `x` and `y`. We can think of this as a vector with its tail at the origin `(0,0)` and the tip at the location given by `(x, y)`



````python
create_canvas(300, 300)
background(0)
x = 120
y = 100
fill(255, 0, 0)
circle(x, y, 5)
stroke(255)
arrow(0, 0, x, y)
show()
````

Many Py5canvas functions work with coordinates written one at the time as in the example above, or with coordinates grouped in pairs of two.



````python
create_canvas(300, 300)
background(0)
x = 120
y = 100
fill(255, 0, 0)
circle((x, y), 5)
stroke(255)
arrow((0, 0), (x, y))
show()
````

This shows how vectors are actually represented in most cases: A sequence of numbers (e.g. `(x, y)`), specifically **two numbers** if the vector &ldquo;lives&rdquo; in a 2d space, three numbers in 3d and so forth. While the vector is encoded as a sequence of numbers, it is useful to still think of it as an arrow in space and as a object of its own.

To take practical advantage of a vector as an object (an arrow) is is useful to use an appropriate Python object. Then we can start using vectors similarly to numbers, enabling operations like addition, subtraction and multiplication by a number (scaling). To enable these functionalities we can use a function to create a vector called `Vector(x, y)`. Let&rsquo;s use vectors to rewrite the previous example using linear interpolation along a line:



````python
create_canvas(400, 200)
background(0)
start = Vector(200, 30)
end = Vector(300, 150)
d = end - start
amt = 0.3
fill(0, 128, 255)
stroke(255)
arrow(start, end)
circle(start + d*amt, 10)
show()
````

In this example we create two vectors `start` and `end` with specific coordinates to indicate two positions on the canvas. We then create a third vector `d` subtracting `start` from `end`. This gives us the vector taking us from `start` to `end`. We can add this vector scaled by a number `amt` to `start` to get a linear interpolation between the two positions. We could have also written `lerp(start, end, amt)` using vectors, which would do exactly the same thing.

If we print one of the vectors we see its x and y components:



````python
start
````

and we can access single components of the vector with the same syntax we use for lists, e.g.



````python
print(start[0], start[1])
````

However, it is usually not necessary to refer to the specific components of a vector except when initializing it using the `Vector` function. Once initialized, we can add up vectors and do operations on them and pass them to drawing functions without worrying too much about the actual values but rather thinking of them visually/geometrically instead. This makes things nicer, since we do not have to worry about getting multiple coordinates right.

Note also that when we print out a vector, we see the word &ldquo;array&rdquo; appearing again. We will cover this more in detail later on, but the reason is that Py5Canvas uses a powerful external library called [NumPy](https://numpy.org) to handle vectors.

### Vector operations



Vectors can be used similarly to numbers but with some specific limitations. The main operations that make visual sense with vectors are **addition**, **subtraction** and **multiplication by a number**. With "visual sense" I mean that you can think about these operations as combinations of arrows, or draw them on paper without too much of an effort.

#### Addition





Adding two vectors creates a new vector, where each component of the new vector is found by adding the matching components of the two original vectors. Think of it like combining two lists of numbers by adding the numbers in the same positions, e.g.



````python
Vector(10, 5) + Vector(5, -4)
````

Visually one way to imagine vector addition is taking one of the vectors and moving its tail to the tip of the other. The addition is the vector connecting the tail of the vector we did not move to the tip of the other:



````python
create_canvas(300, 300)
origin = Vector(0, 0)
a = Vector(50, 50)
b = Vector(100, 30)
translate(100, 100)
stroke(255, 0, 0)
arrow(origin, origin + a)
stroke(0, 255, 0)
arrow(origin, origin + b)
stroke(0)
arrow(origin, origin + a + b)
# Draw the shifted vector
stroke(0, 0, 255, 100)
arrow(a, a + b)
show()
````

#### Subtraction



Similarly to addition, subtracting two vectors creates a new vector, where each component of the new vector is found by subtracting the matching components of the two original vectors:



````python
Vector(10, 5) - Vector(5, 1)
````

If you subtract two vectors `b - a`, think of it as creating a new vector that goes from the tip of `a` to the tip of `b`:



````python
create_canvas(300, 300)
origin = Vector(0, 0)
a = Vector(50, 50)
b = Vector(100, 30)
translate(100, 100)
stroke(255, 0, 0)
arrow(origin, origin + a)
stroke(0, 255, 0)
arrow(origin, origin + b)
stroke(0)
arrow(origin, origin + (b - a))
# Draw the shifted vector
stroke(0, 0, 255, 100)
arrow(a, b)
show()
````

If we think of `a` and `b` as positions, `b - a` gives us the vector that takes us from `a` to `b`:



````python
create_canvas(300, 300)
origin = Vector(0, 0)
a = Vector(50, 50)
b = Vector(100, 30)
translate(100, 100)

fill(255, 0, 0)
circle(a, 5)
fill(0, 255, 0)
circle(b, 5)
arrow(a, b)
show()
````

#### Multiplication by a number



Multiplying a vector by a single number (often referred to as "scalar&rsquo; in more technical texts) simply means scaling the length of the vector while keeping the same direction. If the number is greater than 1, the vector will become longer. If the number is less than 1 it will become smaller, just like using `scale`



````python
create_canvas(300, 300)
origin = Vector(0, 0)
a = Vector(50, 50)
translate(100, 100)
stroke(255, 0, 0)
arrow(origin, origin + a)
stroke(0, 120)
arrow(origin, origin + a*2)
show()
````

You can think of this as **how much of a vector you want**, where the number determines how much.

#### Other operations



Py5canvas has a other functions that make life easier when working with vectors.



##### Built in Py5canvas attributes



Py5canvas has a number of properties that are vectors and make life easier and most of the drawing and transformation functions accept vectors or simply lists of parameters as arguments. For example, we often have seen examples where we translate the origin to the center of the canvas with `translate(width/2, height/2)`. We can also do this with `translate(center)` where `center` is automatically set to be `Vector(width/2, height/2)`.

When using interactive sketches, we can also get the position of the mouse as a vector, using the `mouse_pos` property.



##### Length of a vector



Given a vector we can always query its length using the `mag` function. The word `mag` comes from &ldquo;magnitude&rdquo;, which is a more technical term used for the length of the vector (or &ldquo;norm&rdquo;).



````python
create_canvas(300, 300)
a = Vector(200, 200)
b = Vector(100, 250)
d = b - a
arrow(a, b)
print(mag(d))
show()
````

Dividing a vector by its length, creates a vector of length 1. This is known as &ldquo;normalizing&rdquo; the vector.



##### Orientation of a vector



A vector has a length and also a direction/orientation. For 2d vectors we can get the angle representing this direction using the `heading` function. The functions gives us the orientation of the vector expressed in radians, with the assumption that a horizontal vector pointing to the right has orientation 0:



````python
degrees(heading(d))
````

We can also construct a vector with a given orientation and length 1 using the `direction(angle)` function, which takes an angle in radians as a parameter. For example:



````python
create_canvas(300, 300)
translate(center)
v = direction(radians(45))*100
arrow((0,0), v)
print('Length of vector is', mag(v))
print('Orientation of vector is', degrees(heading(v)))
show()
````

Notice that we multiply the result of direction by 100 so the vector with length 1 becomes a vector with length 100. If we print the actual length and orientation of the vector we get a slightly smaller value (something like the value minus `0.00000000000001`). This is due to the non infinite precision of computers.

In the example above we also avoid using `Vector(0,0)` for the origin, but rather use `(0,0)`. This is fine, but using this syntax will not allow for adding or multiplying elements like we have seen so far.



##### Rotating vectors



It is possibl to rotate a vector using the `rotate_vector(vector, angle)` function. It takes a 2d vector as the first argument and an angle in radians as the second



````python
from py5canvas import *
create_canvas(400, 400)
translate(center)
v = Vector(100, 0)
stroke(0, 128)
for angle in linspace(0, PI*2, 20, False):
    arrow((0, 0), rotate_vector(v, angle))
show()
````

Notice that we added `False` to the end as an argument to the `linspace` function. This avoids drawing the vector with angle 0 twice, since a rotation of `2*PI` is the same as a rotation of 360 degrees and a rotation of 0 degrees.



## Using vectors to draw shapes



We can use vectors to also draw shapes



````python
create_canvas(400, 400)
translate(center)
begin_shape()
for angle in linspace(0, PI*2, 26, False): #Skip last element
    radius = width/3 #random(width*0.4, width*0.5)
    pos = direction(angle)*radius
    vertex(pos)
end_shape(CLOSE)
show()
````

Here we create a circle by doing a `for` loop over a sequence of numbers from zero to `PI` (not included) and then create a position along the circle by scaling the vector `direction(angle)` with length one by multiplying it with the radius. We also introduce `end_shape(CLOSE)` which tells Py5canvas to close the current shape. Try to replace the radius with a random function and to vary the number of points.



### Curve interpolation



Technically when creating a polygon, we are doing linear interpolation between
its points by connecting them with straight lines. Alike P5js and Processing,
Py5canvas allows you to interpolate the points of a polygon with curves, which
gives a smoother looking result. To do so, just replace `vertex` with
`curve_vertex`.



````python
create_canvas(400, 400)
translate(center)
curve_tightness(0.5)
begin_shape()
for angle in linspace(0, PI*2, 16, False): #Skip last element
    radius = random(width*0.2, width*0.5)
    pos = direction(angle)*radius
    curve_vertex(pos)
    #circle(pos, 2)
end_shape(CLOSE)
show()
````

The result is a curve widely used in computer graphics called a
[&ldquo;Catmull-Rom Spline&rdquo;](https://en.wikipedia.org/wiki/Centripetal_Catmull–Rom_spline). It was co-invented, among many other ideas in computer
graphics, by a computer scientist [Ed Catmull](https://en.wikipedia.org/wiki/Edwin_Catmull), who went on to found Pixar.

The `curve_tightness` allows you to control how smooth the curve is, with values near zero being very smooth (zero does not work well) and a value of `1` being like a polygon. You can uncomment the `circle` function above to see the points that are being interpolated by the curve.



## Recap



### Lists



-   Contain a sequence of elements.
-   Intialize with square brackets, e.g. `my_list = [0, 1, 3]`
-   Access or modify elements with index starting from zero
    -   e.g. `my_list[0]` is the first element
-   Access number of elements with `len(my_list)`
-   Add an element `elem` with `my_list.push(elem)`
-   Remove an element from the end with `my_list.pop()`



### Tuples



-   Contain a sequence of elements
-   Initialize with parentheses, e.g. `(0, 1, 3)` or simply comma separated values `0, 1, 3`
-   Cannot be modified once created



### Vectors



-   Sequences of numbers that that describe a direction in space and a length/distance along that direction
    -   And also positions/points
-   Initialize with `v = Vector(x, y)` for the 2d case we will be using most
-   Access/modify elements like lists with square brackets, e.g. `v[0] = 10`.
-   Addition, subtraction and multiplication by a number have a clear visual/geometric meaning.
