## Loops



We have seen how replacing `if` with `while` (and omitting any `elif` or `else`) will **repeat** a block of code **while** the condition is `True`. The automatic repetition of a block of code is called a **&ldquo;loop&rdquo;**, since when the block is ended, execution loops back to beginning of the block, until a certain condition is verified. As an example, when running interactive scripts and writing `run()` at the end of a sketch, the py5canvas engine internally does something similar to this

```Python
def run():
    setup()

    while window_is_open():
        draw()
        show_canvas_in_window()
```
Where the hypotetical function `window_is_open()` returns `True` if the sketch window is open and `False` if you close the window. You define what the `setup` and `draw` functions do, while the `setup` function is automatically called once and the `draw` function is called repeatidly in a loop until you close the window (note that the `window_is_open` and `show_canvas_in_window` functions do not exist and are just here for demonstration puroposes).

The `while` construct is just one way to create loops in Python, and it is typically used in situations where you are creating a repetition that depends on some event happening in the future (e.g. closing a window). However, `while` loops are not recommended in most situations since it is easy to create loops that never exit (e.g. `while True`) and this can lead to the your program hanging.



### For loops



For most situations, you will know how many times you want to repeat code in a loop and for this task (as in many other languages) Python has the `for` loop construct. Say you want to count from `0` to a number stored in a variable called `n`. In Javascript you would do something like:

```Javascript
let n = 10;
for (let i = 0; i < 10; i = i + 1) {
    print(i);
}
```

Firstly, note that in Python you don&rsquo;t have the `i++` syntax ([incrementing](https://www.wordhippo.com/what-is/the-verb-for/increment.html) by one), secondondly we use a quite different but more concise syntax:



````python
n = 10

for i in range(n):
    print(i)
````

As the name implies, by default, the built in `range` function returns a sequence of numbers starting from zero and counting up to (but not including) its argument, so `range(10)` will return a sequence of ten numbers going from `0` to `9`. You can also create [different types of ranges](https://www.w3schools.com/python/ref_func_range.asp) (e.g. counting backwards or in steps) but let&rsquo;s focus on this simple case first.

While it is recommended to use meaningful variable names when writing code, it is conventional to use the `i`, `j`, `k` as variable names in loops. This probably comes from mathematics, where these names are used for indices (for example $x_i$ would be the $i$&rsquo;th of a sequence of $x$&rsquo;s).

So let&rsquo;s say we want to draw a sequence of `n` rectangles in a row of the canvas, we can do:



````python
create_canvas(400, 100)
background(0)
fill(255, 0, 0)

n = 100
spacing = width / n
for i in range(n):
    # print(i)
    rect(i*spacing, 0, spacing*0.95, height)

show()
````

In the example above (if you did not modify it), the variable `i` goes sequentially from `0` until it reaches `4` (that is `n-1`). It consists of `n` numbers, counting from `0` upwards. To fill the canvas we set the left corner of each rect to `i*spacing`. The width of the rectangle is reduced to `spacing*0.8` to leave some space between the rectangle. T

#### 1d loops with randomness



We can combine loops (repetition) with the `random` function to create unpredictable values in a given range. The random functions without parameters `random()` returns a value between zero and one. For example we can use this with a loop and an `if` statement to virtually flip a coin a number of times:



````python
for i in range(10):
    if random() < 0.1:
        print('Heads')
    else:
        print('Tails')
````

Because `random()` returns values between 0 and 1 with equal probability, the epression `random() < 0.5` has the effect of being true 50% of the times. If we reduce the number `0.5` to a smaller number we will have a lower probability of the condition being true.

If we give random **two parameters**, we can control the range of the random numbers. E.g. `random(0, width)` will give a number between `0` and the width of the canvas `width`. The number will never be `width` but it could be a slightly smaller value (e.g. `width-0.0000001`). Let&rsquo;s use thiswith a loop to randomly fill the canvas with shapes, e.g. transparent black circles and with a low probability (5% of the times) color the circles red.



````python
create_canvas(400, 400)
background(255)
for i in range(1600):
    x = random(0, width)
    y = random(0, height)
    radius = random(10, 50)
    if random() < 5/100:
        fill(255, 0, 0)
    else:
        fill(0, 30)
    circle(x, y, radius*0.1)
show()
````

#### 1d Loops to generate curves



When you diplay and image on the screen, it actually consists of a raster (a grid) of many small colored dots (pixels). If the dots are small enough they give the illusion of a smooth image. E.g. zooming into a detail you would have something like this



We can do something similar with curves, for example we can create the illusion of a continuous curve by drawing many dots along the curve. The process of taking points/measurements to approximate something continuous (e.g. a curve or an image) is called &ldquo;sampling&rdquo;. The higher the number of samples, the better our approximation will be. Let&rsquo;s look at an example similar to last week using the `sin` (sine) function and try to modify the `num_samples` variable:



````python
from py5canvas import *
create_canvas(600, 200)
background(0)
stroke(255)
num_samples = 550
for i in range(num_samples):
    x = remap(i, 0, num_samples-1, 0, width)
    y = sin(x*0.05)*(height/3)
    point(x, height/2 + y)
    
show()
````

Here we use the `sin` function that takes a linear sequence of numbers and transforms it into a value continuously oscillating between -1 and 1.

In practice, when we sample a custom defined curve it is better to transform it into a sequence of line segments connecting adjacent dots (samples). To do so we can use a new set of built-in functions: `begin_shape`, `vertex` and `end_shape`.



````python
create_canvas(400, 400)
begin_shape()
vertex(40, 40)
vertex(40 + 100, 40)
vertex(40 + 100, 40 + 100)
vertex(40, 40 + 100)
end_shape()
show()
````

````python
from py5canvas import *
create_canvas(600, 200)
background(0)
stroke(255)
no_fill()
begin_shape()
num_samples = 300
for i in range(num_samples):
    x = remap(i, 0, num_samples-1, 0, width)
    y = sin(x*0.05)*(height/3)
    vertex(x, height/2 + y)
end_shape()
show()
````

The function `begin_shape()` tells Py5canvas that we will start drawing a shape by specifying points along the shape with `vertex(x, y)`. The function `end_shape()` tells Py5canvas that we are done specifying points and to connect them all with line segments. Again, try to modify the `num_samples` variable to see how this works.

You can try to combine the function `sin` with the function `cos`, which is almost identical but shifted version of `sin`, to vary both the X and Y coordinates. This allows us to &ldquo;manually&rdquo; draw a circle:



````python
from py5canvas import *
create_canvas(400, 400)
background(0)
stroke(255)
no_fill()
begin_shape()
num_samples = 100
radius = width/3
for i in range(num_samples):
    t = remap(i, 0, num_samples-1, 0, PI*2)
    x = cos(t*0.5)*radius
    y = sin(t)*radius
    vertex(width/2 + x, height/2 + y)
end_shape()
show()
````

Notice that if we reduce `num_samples` to 5 we get a rotated square. Here we create a variable `t` that remaps the variable `i` given by the `for` loop to a value between `0` and `PI*2`. We will see a bit more about this built in value later with transformations, and even more in future classes, but it is the amount that produces one full oscillation in the sine and cosine functions.

A circle like this is a good place to start when doing experiments, e.g. try multiply `t` inside either `cos` or `sin` by some number, adding another number inside the functions, or even combining multiple `sin` and `cos` with addition and multiplication.



#### Iterating over things

#### Nested `for` loops and grids



Now say we want to create a grid of rectangles, how would we do that? One way to do this is to &ldquo;nest&rdquo; a `for` loop inside another `for` loop, and we can do this by simpy indentig our code e.g.



````python
create_canvas(400, 400)
background(0)
fill(255, 0, 0)

num_rows = 3
num_cols = 3
x_spacing = width / num_cols
y_spacing = height / num_rows

for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        print('i:',i, 'j:',j)
        circle(j*x_spacing + x_spacing/2, 
               i*y_spacing + y_spacing/2, 20) #x_spacing*0.9, y_spacing*0.9)
show()
````

Note that the rectangles are not centered in each tile. We will fix this later using transformations.
We can randomly change a color of a grid cell based on randomness as we did earlier:



````python
create_canvas(400, 400)
background(0)

num_rows = 5
num_cols = 5
x_spacing = width / num_cols
y_spacing = height / num_rows

for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        if random() < 0.2:
            fill(255, 199, 30)
        else:
            fill(255, 0, 0)
        rect(j*x_spacing, i*y_spacing, x_spacing*0.9, y_spacing*0.9)
show()
````

##### Grids



Grids are a useful concept when dealing with computer generated graphics. Computer programs allow us to create procedures that are controlled by parameters and a grid allows us to display different variations of these parameters, e.g if we want to display drawings with different colors in a grid we can use the indices `i` and `j` and `remap` these values to values of color componebts between `0` and `255`:



````python
create_canvas(200, 200)
background(0)
fill(255, 0, 0)
stroke(255)

def my_parametric_drawing(x, y, size, r, g, b):
    radius = size*0.5
    fill(r, g, b)
    circle(x+radius, y+radius, radius)

num_rows = 100
num_cols = 100
x_spacing = width / num_cols
y_spacing = height / num_rows
no_stroke()
for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        r = remap(i, 0, num_rows-1, 0, 255)
        g = remap(j, 0, num_cols-1, 0, 255)
        my_parametric_drawing(j*x_spacing, i*y_spacing, x_spacing, r, 0, g)
show()
````

Grids are useful to visualize combined variations of 2 parameters, but they can also be useful for aesthetic purposes. Indeed, we can see a large use of grids in computer art going from the 60&rsquo;s to this day, since they allow to display different iterations and variations of computer generated shapes in an orderly manner and this orderly look can create a contrast to the use of randomness.

````python
create_canvas(400, 400)
background(0)
no_fill()
stroke(255)

def my_random_drawing(x, y, size):
    radius = size*0.3
    stroke(255, 190)
    begin_shape()
    for i in range(10):
        vertex(x + size*0.5 + random(-radius, radius),
               y + size*0.5 + random(-radius, radius))
    end_shape()

num_rows = 10
num_cols = 10
x_spacing = width / num_cols
y_spacing = height / num_rows
random_seed(4)
for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        my_random_drawing(j*x_spacing, i*y_spacing, x_spacing)
show()
````

## Transformations



In the examples above we had to shift each shape to a position along the grid and make sure that the positions we use we are scaled based on the number of rows and columns we chose. We did so by passing two initial arguments to our functions (`x` and `y`) that specify the position on the grid. While this is practical with simple shapes and code, it can become cumbersome as complexity grows. Enter transformations, which allow us to move, rotate or scale drawing commands as we please.

As we mentioned in previous classes, we draw shapes on the canvas by using a coordinate grid with two axes, one vertical pointing down (the &ldquo;Y&rdquo; axis) and one horizontal pointing to the right (the &ldquo;X&rdquo; axis). In 2d, we specify positions with a pair of numbers called coordinates (often written as a pair of numbers in parentheses, as `(30, 100)`), which define a distance to move along each axis (horizontally and vertically) to reach a point from the position `(0, 0)`.

When you write `create_canvas(...)` or `background` in your code, you start drawing from the top left corner of your canvas. The drawing below shows the canvas in gray and a point relative to the origin, but adds a shift of a few pixels at the beginning of the drawing code by using the `translate` function so we can actually see the axes. This function is a transformation and we will see a few more next. We will use two user defined functions

-   `draw_axes` draws the coordinate axes and displays their length.
-   `draw_point(x, y)` draws a point at the coordinates specified and displays its position.

The functions are defined below and use a particular syntax is the `text` command we have not covered yet, involving [formatted strings](https://www.w3schools.com/python/python_string_formatting.asp) and [keyword (named) arguments](https://www.w3schools.com/python/gloss_python_function_keyword_arguments.asp). Feel free to read the linked articles for more info on how this works, but we will cover these techniques in the future.



````python
def draw_axes():
    x, y = 0, 0
    size = 100
    stroke_weight(3)
    stroke(255, 0, 0)
    arrow((x, y), (x+size, y))
    stroke(0, 255, 0)
    arrow((x, y), (x, y+size))
    fill(0)
    no_stroke()
    circle((x, y), 5)
    fill(0)
    text_size(14)
    text(f"{size}px", x + size/2, y-5, align='center', valign='botton')

def draw_pos(x, y):
    fill(0)
    circle(x, y, 3)
    text(f"({x}, {y})", x, y - 9, align='center')

create_canvas(400, 400)
background(255)
translate(30, 30) # Comment this
draw_pos(0, 0)
fill(0, 40, 90, 40)
rect(0, 0, width, height)
fill(0, 150)
text_size(25)
text('Canvas', [100, 100])
draw_axes()
draw_pos(width/2, height/2)
show()
````

One way of thinking of transformations is to think of operations that move (`translate`), rotate (`rotate`) or scale (`scale`) your current coordinate axes.



### Translations



A translation moves the position of points in space. Think about it as displacement, or viewing yourself from the top, walking around a room, while always facing in the same direction. Or think it adding the first number to all your X coordinates and the second number to all the coordinates.

For example if we write `translate(width/2, height/2)` we can read it as &ldquo;Move the coordinate axes by `width/2` horizontally and `height/2` vertically&rdquo;. Because we start at the top left corner, this will move the &ldquo;origin&rdquo; `(0, 0)` to the center of the canvas:



````python
create_canvas(400, 400)
background(200)
translate(width/4, height/2) # Comment this
fill(0, 150, 250)
circle(0, 0, 150)
square(0, 0, 70)
draw_axes()
draw_pos(0, 0)
show()
````

This could be useful if we want to draw things centered in the canvas, or centered at any position. The idea is do all your drawing code relative to the origin `(0, 0)` and then add `translate` with the desired position before the code. In the code above we could have written `circle(width/2, height/2, 150)` without needing to use `translate`, but drawing code relative to the circle would become more complicated and more difficult to think of thereafter.



### Rotation



As the name implies, rotation points around the current origin. If we have a translation before a rotation, the rotation will happen around the point the translation moved us to. E.g in the example above, let&rsquo;s rotate and draw a square instead of the circle:



````python
create_canvas(300, 300)
background(200)
translate(width/2, height/2) # Comment this
rotate(radians(45))
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
print(radians(45))
````

Notice how our axes are now not oriented straight, but rotated. Rotations happen about the origin and, by default, squares and rectangles are drawn from the top left, so the square is not centered. You can use
`set_rect_mode("center")` to tell Py5canvas that you want to draw from the center instead of the corner (you can use `set_rect_mode("corner")` to revert).

In the code above you can see that we use the function `radians` inside of the `rotate` function. Radians are a way to measure angles where we measure the distance along a circle with radius $1$. The distance along half a circle is the [very celebrated](https://www.piday.org) number usually referred to with the greek letter [$\pi$](https://en.wikipedia.org/wiki/Pi) or `PI` in P5js and Py5Canvas. It is an irrational number, that it is it has an infinte number of decimals with no repreating pairs. Tha value of $\pi$ is:
$$
3.141592653589793238462643383279502884197...
$$
and goes on forever. It corresponds to `180` degrees, which is a little easier to remember. A full rotation of $360$ degrees corresponds two times $\pi$, that is `2*PI`. For the case above, we could have written `rotate(PI/2)`, which corresponds to a clockwise roation of $45$ degrees.



````python
print(radians(45))
````

You may wander why not just use degrees. The reason is that the mathematical functions used to calculate roations (e.g `sin` and `cos`) operate on radians and relate to circles with radius 1.



### Scaling



Scaling enlarges or shrinks shapes and the coordinate axes by a specifed amount. An amount smaller than one (`<1`) shrinks, an amount greater than one (`>1`) enlarges, and `1` does nothing. Imagine multiplying all your coordinates by that number.



````python
create_canvas(400, 400)
rect_mode('center')
background(200)
translate(width/2, height/2) # Comment this
rotate(radians(45))
scale(0.5)
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

Notice that if we applied a scale `< 1` in the code, the coordinate axes have shrinked. That means that our &ldquo;ruler&rdquo; has shrinked, so a distance of 100 is less than 100 pixels now.
The example above shows a **uniform** scaling, it scales X and Y coordinates by the same amount. If we specify two numbers we can scale the horizontal and vertical coordinates by different amounts:



````python
create_canvas(400, 400)
rect_mode('center')
background(255)
translate(width/2, height/2) # Comment this
rotate(radians(45))
scale(0.5, 1)
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

Notice that the rectangle is scaled along the coordinate axes **transformed** by the previous translation and rotation.
If we apply the non-uniform scale **before** the translation, we will get a more unexpected result:



````python
create_canvas(400, 400)
rect_mode('center')
background(255)
translate(width/2, height/2) # Comment this
scale(0.25, 1)
rotate(radians(45))
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

This results in a transformation known as a &ldquo;shear&rdquo;, which also modifies angles in addition to coordinates. We will not cover this in detail during this module, but in practice, the rotation has a different effect because once we scaled non-uniformely circles are not circles become ellipses and rotations behave differently.



### Combining transformations - Order matters!

### Thinking of transformations, first person (local) view



The way in which we conceptualized tranformations above, can be compared to a &ldquo;first person view&rdquo;, where we are in a 2d world and viewed from the top. If we put our left arm forward and our right arm sideways, our right left forward arm is the X axis (the red axis) and our right arm is the Y axis (the green axis) and our head is the origin (the black dot). If we move (translate) two step2 forward and then one step to the right, we did `translate(2, 1)` but the origin is still at our head. If we rotate a positive amount (say 45 degrees), we rotate towards the right arm. Our arms remain perpendicular and the left arm always faces forward. Things get weird if we apply a uniform scale, as we either become a dwarf or a giant. Then one step becomes either shorter or longer depending on the scale.

This way of thinking of transformations is perhaps the most intuitive unless we introduce non-uniform scaling. It can make building articulated bodies easy, for example we can build a &ldquo;robotic arm&rdquo; made of many links.



````python
def draw_link(length):
    stroke(0)
    no_fill()
    rect(0, -30, length, 60)
    draw_axes()

create_canvas(400, 400)
background(255)
translate(30, height/2)
for i in range(5):
    draw_link(100)
    translate(100, 0)
    rotate(radians(40))
    scale(0.8)
show()
````

#### Thinking of transformations, god (global) view



Anther way to think of transformations is to think of manipulating objects on the canvas while the coordinate system and the origin remain fixed. Think of it as moving or rotating paper on a board when making a collage. The main drawback of this conceptualization is that you must think of the rder of transformations in &ldquo;reverse&rdquo;, starting from the objects/shapes we are drawing and going backwards.
This makes more sense if you think of transformations as **functions** that applied to what follows and return a transformed shape.  E.g. earlier we used a function inside a function `print(radians(45))`, and we read this backwards: The function ~radians~ is applied to the value 45 and the function ~print~ prints the result. With a composition of transformations of this type (these examples are "pseudocode" so they will not work):
```python
translate(...)
scale(...)
rotate(...)
draw_shape()
```
you can imagine this to be something equivalent to
```python
translate(scale(rotate(shape)))
```

where `rotate(shape)` gives a new rotated shape and `scale(rotate(shape))` scales the rotated shape, etc... So you read the transformations backwards, going from draw shape, and each transformation is applied to everything that follows. 

One example where this conceptualization is useful is modeling a planetary system with the sun at the center and one (for now) planet revolving around it (use a sketch to make that happen). We will think of it as follows. We first move our origin to the center of the canvas (that is the firt person view described earlier) and draw the sun at the origin. Now let&rsquo;s take a god&rsquo;s point of view and think of how we would place the planet with respect to the sun and origin:

-   We shift the planet from the origin.
-   We rotate the shifted planet around the origin by a certain amount.

To do this we write:



````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)
# orbit
no_fill()
stroke(255, 30)
circle(0, 0, 150)
no_stroke()
# Planet
rotate(radians(30)) #PI/2)
translate(150, 0)
fill(0, 150, 200)
circle(0, 0, 20)
show()
````

E.g. if you add another planet to the previous canvas without somehow resetting your transformation (note that in a notebook we can keep on working on the same canvas)



````python
rotate(-PI/2)
translate(100, 0)
fill(0, 180, 100)
circle(0, 0, 10)
show()
````

We can avoid these problems by using a construct known as the &ldquo;trasformation stack&rdquo;, which allows you to save (push) the current trasformation and restore (pop) the last saved transformation using the `push_matrix()` and `pop_matrix()` commands. The name &ldquo;matrix&rdquo; appears because transformations are usually encoded as a matrix (a grid) of numbers.

So to draw an arbitrarily large number of planets we can do:



````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)

for i in range(5):
    dist = 70 + i*30
    no_fill()
    stroke(255, 30)
    circle(0, 0, dist)
    no_stroke()
    push() # Save current transformation
    rotate(random(-PI, PI))
    translate(dist, 0)
    fill(0, 150, 200)
    circle(0, 0, random(5, 10))
    pop() # Restore last saved transformation
show()
save('planets.svg')
````

An alternative and slightly more advanced **syntax** that you can use in Python and will not work in Javascript is the following:



````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)

for i in range(5):
    dist = 70 + i*30
    no_fill()
    stroke(255, 30)
    circle(0, 0, dist)
    no_stroke()
    with push_matrix(): # Save current transformation
        rotate(random(-PI, PI))
        translate(dist, 0)
        fill(0, 150, 200)
        circle(0, 0, random(5, 10))
show()
````

Here we use the indentation syntax, together with the keyword `width` to tell
the program that after all the indented code following `with push_matrix():`, we need to automatically restore the previous transformation with `pop_matrix()`.



### Back to grids



In the earlier examples, we have seen that we needed to carefully adjust coordinates to place our drawings in a grid.
With the tools we have learned we can now simplify the process, and just think of a drawing with respect to the origin and then transform it for each cell of a grid. Let&rsquo;s start by working on what we want to display in each cell and put it into a function.

We will work in a 300x300 canvas with center at the origin:



````python
tile_size = 300
create_canvas(tile_size, tile_size)
def draw_tile():
    rect_mode('center')
    no_stroke()
    fill(50)
    square(0, 0, 250)
    fill(255, 128)
    for i in range(30):
        circle(random(-100, 100),
               random(-100, 100),
               20)

translate(width/2, height/2)
draw_tile()
save('myimage.svg')
show()
````

Because the `draw_tile()` function uses randomness, it will be different each time we use it.
Now to the grid code, it is similar to the one we did previously, but we will will use `push_matrix` and `pop_matrix` to transform the same drawing function without needing to supply the coordinates:



````python
create_canvas(500, 500)

num_rows = 10
num_cols = 10
x_spacing = width / num_cols
y_spacing = height / num_rows

for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        push_matrix()
        # Move to the tile top left corner
        translate(j*x_spacing, i*y_spacing)
        # Move to the center of the tile
        translate(x_spacing/2, y_spacing/2)
        # Scale to fit the grid
        scale(x_spacing/tile_size)
        # and draw
        draw_tile()
        pop_matrix()
show()
````

Because we put the transformation between `push_matrix()` and `pop_matrix()` each transformation works independently and does not &ldquo;add up&rdquo; to the previous. Note that we originally designed a random tile on a 300x300 (the 300 is stored in the `tile_size` variable) canvas with the origin shifted to the center. So our transformation code first moves the origin to the top left corner of the tile, then moves further to the center of the tile, and then scales the coordinates so that they fit in the (smaller) size of a tile. The amount of scaling is given by `x_spacing/tile_size`, where the scale will be smaller than 1 if `tile_size > x_spacing`.
