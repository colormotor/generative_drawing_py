## Variables
Python is a dynamically typed language, which means that you do not have to declare the type of a variable when you create it. You can simply assign a value to a variable name and Python will figure out the type for you. 

You can query the type of a variable using the `type` function:

````python
a = 'some text'
type(a)
````

## Conditionals
Python boolean values are `True` and `False` (NOTICE THE CAPITALS!). You can use them in conditionals using `if`, `elif` and `else` statements. Here is an example:

````python
a = 5

if a==10:
    print("A is 10")
elif a==5:
    print("A is 5")
else:
    print("A is something else:", a)
````

This shows one big difference between Python and JavaScript: in Python indentation matters and we don't use curly brackets to define blocks of code. 
Each statement, such as an `if` is followed by a colon `:` and a new line. The code that would be inside curly brackets in Javascript is indented. The standard is to use 4 spaces for each indentation level but most editors will convert "tabs" to spaces automatically.

Think of it as nicely formatted JavaScript, but without the curly brackets.

Also note that in Python we use `elif` instead of `else if`.


## Functions and indentation
The biggest difference between Javascript and Python is the way in which we define code bblocks. In Javascript, we define the
body of a function, or the operations following an `if` statement, with curly
braces (`{` and `}`). The curly braces define what is known as &ldquo;scope&rdquo;.

    function doSomething() {
        if (a == 10){
            print("A is 10");
        }else if (a==5){
            print("A is 5");
        }else{
            print("A is something else")
        }
    }


Let's look at the same function in Python:

````python
def do_something():
    if a==10:
        print("A is 10")
    elif a==5:
        print("A is 5")
    else:
        print("A is something else")
````

Here we used "snake_case" as opposed to "camelCase" in Javascript. This is a common convention in Python, but not mandatory. Most importantly:

1. We start with the keyword `def`
2. This is followed by the function's name
3. And a set of comma-separated names **inside parentheses**.
4. Followed by a colon `:` and a new line
5. The **indented** code under the function declaration is what the function does.

As with conditionals, the scope of a function, or any other code block, is defined by indentation. 



Let's look at a practical example:

````python
def my_shape(x, y, w):
    circle(x, y, w)
    rect_mode(CENTER)
    rect(x, y, w/2, w/2)

create_canvas(500, 300)
x = width/2
my_shape(x, 100, 60)
my_shape(x, 150, 30)
my_shape(x, 200, 15)
show()
````

### Globals
In Python, variables defined outside of functions are known as "global" variables. They can be accessed from within functions, but if you want to modify them inside a function, you need to declare them as `global` within that function. This differs from Javascript, where global variables can be modified directly within functions without any special declaration. You might need to use this in interactive sketch that for instance keep track of time or other data stored across frames.


E.g. let's define two global variables:

````python
value1 = 5
value2 = 10
````

And then increment these each time you call a function:

````python
def increment():
    #global value1, value2
    value1 = value1 + 1
    value2 = value2 + 1
    print(value1, value2)

increment()
````

This will only work if you uncomment the line with `global` inside the function. The comma separated variables after `global` are the ones we want to modify inside the function.

## Loops 
In Python, we usually loop by **iterating** over a **sequence of values**.

A Python sequence is an ordered collection of items (variables) that can be accessed by an index. The most common sequences in Python are "strings", "lists", "ranges", "tuples". 

- A "string" (type `str`) is a sequences of characters defined by surrounding text with quotes, e.g. `"hello"` or `'hello'`. 

- A "list" (type `list`) is a sequence of variables. Items in a list can be modified, removed or added. 
  A list can be defined as a comma separated sequence of variables enclosed by square brackets, 
  e.g. `["apple", "banana", "milk", "bread"]`.
 
- A "range" (type `range`) is a sequence of integers, it is similar to a list but it is faster when accessing one item after the other.
  E.g. `range(10)` gives us a sequence of ten numbers from `0` to `9` (note that we start counting from zero)

- A "tuple" (type `tuple`) is a sequence of variables similar to a list, but its elements cannot be modified.
  It can be defined as a comma separated sequence of variables, optionally enclosed by parentheses.
  E.g. `(0, 1)` or `x, y`.

In Python, you do loops by iterating over a sequence. Intuitively, imagine a repetitive task such as scanning the items in your basket at the till of a supermarket. You would take each item one by one and scan it. You can think of it as:

    For each item in the sequence, do something with the item.

In Python it could look something like:
```python
basket = ["apple", "banana", "milk", "bread"]
for item in basket:
    scan(item)
```

You can use the same procedure with any sequence-like object in Python, e.g we could count the number of characters in a string with a `for` loop:

````python
txt = "Hello world"
count = 0
for c in txt:
    print(count, c)
    count += 1 #+= 1 # This is a shorthand for i = i + 1
print(count, 'characters')
````

We can access the elements of sequences using the `[]` operator and a zero-based index, just like with arrays in Javascript. E.g.:

````python
txt[0]
````

But also backwards with negative indices (starting from -1 for the last element):

````python
txt[-1]
````

### Counting

So far we have explicitly defined the items in a sequence, e.g. the items in the `basket` list or the characters in a string. But what if we want to count from 0 to 10? 

We could define a list with the numbers from 0 to 10, but that would be inconvenient 

````python
for i in [0, 1, 2, 3, 4, 5, 6, 7, 8]:
    print(i)
````

Instead, we can use a built in `range` type, which generates a sequence of numbers for us. In its most basic form, the `range` function takes one argument and it will generate a sequence of numbers from 0 up to (but not including) that number: 

````python
for i in range(10):
    print(i)
    
````

You can imaging it similarly to the string we saw earlier, but with numbers:



Note, that while it is recommended to use meaningful variable names when writing code, it is conventional to use the `i`, `j`, `k` as variable names for indices into sequences or when counting as we did above.

You can also create [different types of ranges](https://www.w3schools.com/python/ref_func_range.asp) (e.g. counting backwards or in steps) but let&rsquo;s focus on this simple case first.

Let's look at an example of the previous drawing using `range` to draw an arbitrary number of shapes:

````python
create_canvas(400, 300)
background(255)
fill(200, 100) # Two values means "white intensity, opacity"
x = width/2
n = 100 # Number of shapes
w = 120
y = 100
for i in range(n):
    circle(x, y, w)
    y += 5 # Increase y position
    w *= 0.9 # Decrease radius (90%) 
show()
````

### Remap
In P5js it is useful to use the `map` function to map one range of values to another. E.g. mapping a value from the range 0-100 to the range 0-1 or to positions along the width of the canvas. Because `map` is a built-in Python function, in Py5canvas we use `remap` instead. It works the same way:


````python
create_canvas(400, 400)
background(0)
n = 50
for i in range(n):
    fill(remap(i, 0, n-1, 0, 255)) # remap i to a number between 0 and 255
    rectangle(remap(i, 0, n, 0, width),
              0,
              20,
              random(0, height))
show()
````

### Grids
So far we have seen single 

We can use indentation to created nested loops. What in Javascript would be:
```Javascript
for (let i=0; i<10; i++){
    for (let j=0; j<10; j++){
        ...
    }
}
```
Becomes in Python:
```python
for i in range(10):
    for j in range(10):
        ...
```

Let's use this to create a grid of shapes:

````python
create_canvas(400, 400)
background(0)

num_rows = 5
num_cols = 5
x_spacing = width / num_cols
y_spacing = height / num_rows

for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        x = remap(j, 0, num_cols, 0, width)
        y = remap(i, 0, num_rows, 0, height)
        rect(x, y, x_spacing, y_spacing) 
show()
````

And here is a version that creates random shapes for each cell in the grid. Here we use `push_matrix` and `pop_matrix` to transform the drawings for each cell:

````python
create_canvas(400, 400)
background(255)
no_fill()
stroke(0)

def my_random_drawing(w, h):
    radius = min(w, h)*0.3 
    begin_shape()
    for i in range(10):
        vertex(w*0.5 + random(-radius, radius),
               h*0.5 + random(-radius, radius))
    end_shape()

num_rows = 10
num_cols = 10
x_spacing = width / num_cols
y_spacing = height / num_rows
random_seed(4)
for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        push_matrix()
        translate(j*x_spacing, 
                  i*y_spacing)
        my_random_drawing(x_spacing, y_spacing)
        pop_matrix()
show()
````

# Raster/vector graphics and drawing machines

## Raster graphics



Images on the computer screen are approximated by an arrangement of colored pixels over a grid (raster). The color of each pixel can be described with a red, a green and a blue component (RGB).

E.g. when writing `circle` or other instructions in Py5Canvas, what we see on our screen is just a approximation made of “little squares”. 

## Vector graphics



The instructions we write in a Py5Canvas program, are mostly **vector graphics** instructions. 
We specify shapes through their parameters, e.g.:
- A circle can be defined with it’s center (x, y) coordinates and its radius or diameter
- A line with a pair of coordinates (x1, y1) and (x2, y2)
And so forth... 




When these shapes are displayed, they are converted to pixels with a process known as "rasterization" or raster-conversion. With Py5canvas this is taken care of by vector graphics library called [“Cairo”](https://www.cairographics.org).

These shapes are defined mathematically and can be scaled to any size without loss of quality. Their definition also contains information on how to reproduce them with a machine, such as a plotter, a laser cutter or a CNC milling machine.



Not all primitives in Py5canvas are vector graphics, e.g. images loaded with `load_image()` are raster graphics and simply define what pixels appear on the screen. E.g. last week we did:
```Python
def setup():
    create_canvas(600, 600)
    background(0)

def draw():
    t = frame_count
    copy(0, 0, width, height, 0, 1, width, height)  # scroll down
    stroke(255)
    ...
```

The `copy` command copies **pixels** from one part of the canvas to another. This is a raster graphics operation and not a vector graphics operation and you would not be able to draw anything of the output with a plotter. 

The same holds for drawing an image, where the pixels of the image are copied to the canvas and there is no information for a plotter to draw anything meaningful.

 Drawing these kind of output with a drawing machine, would require the inverse process of rasterization: converting pixels to vector shapes, a process known as "vectorization" or "image tracing". This is a complex process and not always possible to do with good results,
 
 

 but we will look at (simple) examples in future weeks.

## Turtle graphics
Another fun thing to explore with the plotter is to use the turtle graphics module in Py5canvas. This does not exist in P5js but it is easy to learn. You can find a an interactive example in the script [turtlegraphics.py](./turtlegraphics.py) or in a notebook cell:


````python
from py5canvas.turtle import Turtle 

create_canvas(600, 600)
background(255)
translate(width/2, height/2)
t = Turtle()
for i in range(36):
    t.right(10)
    t.circle(120, steps=11)
stroke(0)
stroke_weight(1)
no_fill()
shape(t.paths) # This d
show()
save('turtle_drawing.svg')
````
