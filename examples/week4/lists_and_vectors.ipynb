## Sequences: lists and tuples



So far we have seen how to use Python sequences to store multiple values and enable iteration. ### Lists



Lists in Python are similar to JavaScript&rsquo;s `Array` object, which you might have
seen already. Lists are like &ldquo;containers&rdquo; that allow you to store a sequence of
multiple values into a single variable. The word &ldquo;sequence&rdquo; is important here
because the items in a list appear in a specific order, so you have a first item,
a second item and so forth.

Let&rsquo;s create a simple list, by using the &ldquo;square bracket notation&rdquo;



````python
my_list = ['first', 'second', 'third', 'fourth']
print(my_list)
````

Here we create a list of items by placing the items within square brackets and
separating them with commas.

The list has a number of elements or size, which can be queried using the Python
built-in `len` function:



````python
len(my_list)
````

#### Accessing list elements



Single elements can be accessed using an &ldquo;index&rdquo; in the list, starting from
zero. For example



````python
my_list[0]
````

will give us the first element in the list.

Remember, **to access an element of a list we use the variable name followed by
square brackets containing the index**. Notice that, similarly to JavaScript and
similarly to the way in which we constructed loops, indices start from zero. As
a result the last element of a list can be accessed with the index
`len(my_list)-1`



````python
my_list[len(my_list)-1]
````

##### Extra info



Writing `len(my_list)` everytime you want to access the end of a list can be annoying. Python helps us save precious energy when typing, so we can simply use negative indices and access the list &ldquo;backwards&rdquo;. As an example, the last element can be also accessed with



````python
my_list[-1]
````

When you see a negative index within square brackets, imagine putting the length of the list before the negative sign.



#### Setting elements



You can use the index syntax to **set** elements of a list, for example



````python
my_list[0] = 100
print(my_list)
````

Will set the first element of a list to be `100`.



#### Lists methods and Python &ldquo;objects&rdquo;



Before we go on we need to touch on the concept of an &ldquo;object&rdquo; in programming.

What is an object? In Python, almost every variable you create, including lists,
is an &ldquo;object&rdquo;. It is a fancy way of saying a &ldquo;thing&rdquo; that has some attributes and can perform certain actions. Objects have:

-   Attributes (information about them)
-   Methods (things they can do, functions specific to the object)

The way in which we access attributes and methods is the &ldquo;dot notation&rdquo;. E.g. if you had an object representing a virtual ball, something like

    ball.bounce()

will perform the `bounce` function, specific to the specific **instance** of the object that the variable `ball` refers to.

A function like this one that is linked to a specific object function is often referred to as a method. The "dot" indicates that the method is attached to object on the left of the dot.

For attributes, we have a similar notation without the parentheses, so

    c = ball.color

will access the value of the attribute `color` specific to the `ball` object and place it into a new variable called `c`.

##### Dynamically manipulating lists

The cell prints out the value because `pop` returns the last element and a notebook cell always prints the result of the last instruction. 


Now the list is one element shorter

````python
my_list.pop(0)
````

### Tuples



Another common way to describe sequences of items in Python is the `tuple`.
Tuples behave similarly to lists but **do not allow modification**. In technical
jargon we say that tuples are *immutable*.

One of the practical utilities of tuples is to organize data, e.g a 2D point (x,y) or a RGB color (r,g,b). To create a tuple you replace the square brackets with parentheses:

````python
my_tuple = ('first', 'second', 'third')
````

Then we can access the elements or the length of the tuple similarly to lists. So

````python
print(my_tuple[0])
print(len(my_tuple))
````

However, now we are not able to modify an element of the tuple, so this will throw an error:



````python
my_tuple[0] = 100
````

Tuples can be initialized without parenthes, making them an intrinsic part of the Python syntax.
For example the above could have been:



````python
my_tuple = 'first', 'second', 'third'
````

This is particularly useful since we can make multiple assignments in one line using this method:



````python
r, g, b = 10, 20, 30
````

Where each element on the left of the `=` is being assigned the corresponding element on the right. This can allow you to quicly swap two values, e.g



````python
r, g = g, r
````

Where `g` gets assigned to `r` and `r` gets assigned to `g`. Doing this without this syntax would require you to use a temporary variable to store one of the two values you want to swap.

We can also &ldquo;distribute&rdquo; values from a tuple into multiple variables, e.g



````python
rgb = (255, 100, 0)
r, g, b = rgb
````

where we first create a tuple called `rgb` with three values, and then assign the values in order to three variables called `r`, `g` and `b`.



We can use this kind of syntax in loops to access the elements of a tuple in a sequence. As a practical example let's write a loop that creates three randomly positioned circles, while assigning each circle a color of choice:

````python
create_canvas(300, 300)
background(0)
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
for r, g, b in colors:
    fill(r, g, b)
    circle(random(0, width), random(0, height), 100)
show()
````

Many of the functions in Py5canvas accept tuples or other kind of sequences as arguments. For example, the `fill` function can accept a tuple of three values representing the RGB color and the `circle` function can accept a tuple of two values representing the (x,y) position of the circle.

````python
create_canvas(300, 300)
background(0)
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
for clr in colors:
    fill(clr)
    circle((random(0, width), random(0, height)), 100)
show()
````

## Linear interpolation and vectors



In earlier sessions, we have seen the `remap` function, which acts as a "translator": it takes 
a value in one range and places it proportionally into another range. One
useful analogy for this is a slider in a user interface. In that case the slider, the input is a horizontal position
along the slider, while the output is a number between the minimum and maximum value of the slider.

Internally this remapping procedure relies on a procedure called &ldquo;linear
interpolation&rdquo; (often referred to as &ldquo;lerp&rdquo;).



### Lerp



Linear interpolation is the process of finding points between two or more
points. Linear interpolation finds a position between a starting and an ending
point as if there were a line connecting them. This can be done by specifying
the relative &ldquo;percentage&rdquo; between the two points expressed as a value between 0
and 1 (with 1 = 100%), then linear interpolation is simply

```python
start + (end - start)*percentage
```

We can do this manually, or use the function `lerp(start, end, amt)` where `amt`
is the percentage. Think of it as a shorthand for `remap(amt, 0, 1, start, end)`.

For example let&rsquo;s draw a circle on a position along a
horizontal line with a circle at a certain percentage along the line:



````python
create_canvas(400, 100)
translate(width/2, height/2)
start = -100
end = 100
amt = 0.9
line(start, 0, end, 0)
circle(lerp(start, end, amt), 0, 10)
show()
````

Try modifying the value of the `amt` variable. If the value of `amt` is less or
greater than 1 then the circle is not on the line segment anymore, which
technically is called &ldquo;extrapolation&rdquo;.

You can check an interactive version of this example [here](./slider.py).

We actually have seen an example of using `lerp` (written manually) in the drawing example of Week 3. There we were find a line between the current &ldquo;brush&rdquo; position and the current mouse position and using linear interpolation to move a small amount along that line. More on that later in this document.



## Vectors



In the examples above we saw how to use linear interpolation to find a point along a horizontal line. We were referring to the values on the horizontal as &ldquo;points&rdquo;, but what if we wanted to find a point along a line that is not horizontal? 
If we think of 2d coordinates we could do this with **two linear interpolations**, one for the X coordinate and one for the Y, e.g.

````python
create_canvas(400, 200)
start_x = 100
start_y = 30
end_x = 300
end_y = 150
amt = 0.3
arrow(start_x, start_y, end_x, end_y)
circle(lerp(start_x, end_x, amt),
       lerp(start_y, end_y, amt),
       10)
show()
````

Here we use `arrow` instead of `line` to indicate the direction of movement.

In practice, this arrow is called a **vector**. Think of it as an arrow in space (in our case 2d) that has a length and a direction. It encodes the **distance** and the direction of the displacement from its tail to its tip. 

In 2d you can think of something like this:



In practice, we have been working with vectors all along. For example, take the position of a point on the canvas with coordinates `x` and `y`. We can think of this as a vector with its tail at the origin `(0,0)` and the tip at the location given by `(x, y)`



````python
create_canvas(300, 300)
x = 120
y = 100
arrow(0, 0, x, y)
show()
````

Many Py5canvas functions work with coordinates written one at the time as in the example above, or with coordinates grouped in pairs of two.



````python
create_canvas(300, 300)
x = 120
y = 100
circle((x, y), 5)
arrow((0, 0), (x, y))
show()
````

This shows the most common representation of vectors: a sequence of numbers. 

On our **2D** canvas, vectors (arrows) are represented as **two numbers** (e.g. `(x, y)`). In 3d these would be three numbers and so forth. 

While we encode a vector as a sequence of numbers, it is useful to still think geometrically as an arrow in space rather than numbers (coordinates).
It is also useful to think of vectors as arrows with a tail that can be located anywhere in space, not just at the origin.

To make a practical advantage of this representation, we need to encode vectors not as a tuple or list, but as a specific Python object. 

Then we can start using vectors similarly to numbers, enabling operations like addition, subtraction and multiplication by a number (scaling). 
To enable these functionalities we can use a function that creates a vector called `vector(x, y)`. 

E.g. 

````python
vector(3, 5) + vector(4, 2)
````

Using tuples (or lists) this would result in a concatenation of elements, which is useful in its own right but not what we want when working on 2d graphics:

````python
(3, 5) + (4, 2)
````

Let's look at geometric examples of these operations:

### Vector operations



Vectors can be used similarly to numbers but with some specific limitations:
The main operations that make geometric (visual) sense with vectors are **addition**, **subtraction** and **multiplication by a number**. 
With "visual sense" I mean that you can think about these operations as combinations of arrows, or draw them on paper without too much of an effort (at least for the 2d case!).

#### Addition





Adding two vectors creates a new vector, where each component of the new vector is found by adding the matching components of the two original vectors. Think of it like combining two lists of numbers by adding the numbers in the same positions, e.g.



````python
vector(10, 5) + vector(0, 1)
````

Visually one way to imagine vector addition is taking one of the vectors and moving its tail to the tip of the other. 
The resulting vector goes from the tail of the vector we did not move to the tip of the one we moved:

````python
def draw_vector(v, origin=(0, 0)):
    arrow(origin, origin+v)

create_canvas(300, 300)
a = vector(50, 50)
b = vector(100, 30)
v = a + b
translate((100, 100)) # Move origin 
stroke(255, 0, 0)
draw_vector(a)
stroke(0, 255, 0)
draw_vector(b)
stroke(0)
draw_vector(v)
show()
````

>NOTE the `draw_vector` function above has a "default argument" called `origin`. By default it takes the value we assinged it within the parentheses. However, we can override this by specifying its value when calling the function. This can be done by either adding an argument to the function call (e.g `draw_vector(a, (10,10)))` or specifying the argument by name (e.g. `draw_vector(a, origin=(10, 10))`)

#### Subtraction



Similarly to addition, subtracting two vectors creates a new vector, where each component of the new vector is found by subtracting the matching components of the two original vectors:

````python
vector(10, 5) - vector(5, 1)
````

If you subtract two vectors `b - a`, think of it as creating a new vector that goes from the tip of `a` to the tip of `b`:



````python
create_canvas(300, 300)
a = vector(50, 50)
b = vector(100, 30)
v = b - a
translate((100, 100)) # Move origin 
stroke(255, 0, 0)
draw_vector(a)
stroke(0, 255, 0)
draw_vector(b)
stroke(0)
draw_vector(v, a)
show()
````

If `a` and `b` are treated as positions, `b - a` gives us the vector that takes us from `a` to `b`:

````python
create_canvas(300, 300)
origin = vector(0, 0)
a = vector(10, 10)
b = vector(100, 130)
translate(100, 100)

fill(255, 0, 0)
circle(a, 10)
fill(0, 255, 0)
circle(b, 10)
arrow(a, b)
show()
````

#### Multiplication by a number



Multiplying a vector by a single number (often referred to as "scalar&rsquo; in more technical texts) simply means scaling the length of the vector while keeping the same direction. If the number is greater than 1, the vector will become longer. If the number is less than 1 it will become smaller, just like using `scale`





````python
create_canvas(300, 300)
a = vector(50, 50)
translate(100, 100)
stroke(255, 0, 0)
draw_vector(a)
stroke(0, 120)
draw_vector(a*2)
show()
````

You can think of this as **the amount of the vector that you want**, where the number determines the amount.

## Vectors and coordinate systems
Indeed, when we define coordinates we can consider our coordinate axes as vectors as well. For example, when we create a new canvas, we start with two coordinate axes that are two perpendicular vectors of length 1 (unit vectors) pointing along the X and Y directions. These vectors are commonly referred to as the basis vectors (they form the basis for the coordinate system). Then, any point is a weighted sum of these two vectors. For example the point (100, 50) can be thought as 100 times the x axis vector plus 50 times the y axis vector.  Zooming into the canvas we would get:

````python
create_canvas(400, 400)
translate(width/2, height/2)
zoom = 50
xaxis = vector(1, 0)*zoom
yaxis = vector(0, 1)*zoom
x, y = 1.5, 3
pos = x*xaxis + y*yaxis

stroke('#ff0000') # Using web color notation
draw_vector(xaxis)
stroke('#ff0000', 90)
draw_vector(xaxis*x)
stroke('#00ff00')
draw_vector(yaxis)
stroke('#00ff00', 90)
draw_vector(yaxis*y)
stroke('#0000ff', 90)
draw_vector(pos)

fill(0)
no_stroke()

circle(pos, 10)
stroke(0, 50)
line(x*xaxis, pos)
line(y*yaxis, pos)

show()
````

In the example above we multiplied our x and y axis vectors by a zoom factor, because otherwise the vectors of length one would be too small to see! This operation is exactly what happens when we use the `scale` function, which scales the coordinate system by multiplying the basis vectors (the axes) by the scaling factor. Similarly if we were to rotate the two axis vectors by the same angle, we would be rotating the coordinate system identically to using the `rotate` function.



Indeed, in practice transformations are usually represented as a "matrix" (hence `push_matrix()`) of basis vectors stacked in its columns. ##### Length of a vector



Given a vector we can always query its length using the `mag` function. The word `mag` comes from &ldquo;magnitude&rdquo;, which is a more technical term used for the length of the vector (or &ldquo;norm&rdquo;).



````python
create_canvas(300, 300)
a = vector(200, 200)
b = vector(100, 250)
d = b - a
arrow(a, b)
print(mag(d))
show()
````

Dividing a vector by its length, creates a vector of length 1 (a unit vector). This is known as &ldquo;normalizing&rdquo; the vector. In 2D, the length of the vector with components $x$ and $y$ is given by Pythagora's formula 

$\sqrt{x^2 + y^2}$

and for higher dimensions you simply adding all the square vector components inside the square root.



##### Orientation of a vector



A vector has a length and also a direction/orientation. For 2d vectors we can get the angle representing this direction using the `heading` function. The functions gives us the orientation of the vector expressed in radians, with the assumption that a horizontal vector pointing to the right has orientation 0:



````python
degrees(heading(d))
````

We can also construct a vector with a given orientation and length 1 using the `direction(angle)` function, which takes an angle in radians as a parameter. For example:



````python
create_canvas(300, 300)
translate(center)
v = direction(radians(45))*100
arrow((0,0), v)
print('Length of vector is', mag(v))
print('Orientation of vector is', degrees(heading(v)))
show()
````

Notice that we multiply the result of direction by 100 so the vector with length 1 becomes a vector with length 100. If we print the actual length and orientation of the vector we get a slightly smaller value (something like the value minus `0.00000000000001`). This is due to the non infinite precision of computers.

In the example above we also avoid using `vector(0,0)` for the origin, but rather use `(0,0)`. This is fine, but using this syntax will not allow for adding or multiplying elements like we have seen so far.



##### Rotating vectors



It is possibl to rotate a vector using the `rotate_vector(vector, angle)` function. It takes a 2d vector as the first argument and an angle in radians as the second



````python
from py5canvas import *
create_canvas(400, 400)
translate(center)
v = vector(100, 0)
stroke(0, 128)
n = 20
for i in range(n):
    angle = remap(i, 0, n, 0, TWO_PI)
    arrow((0, 0), rotate_vector(v, angle))
show()
````

## Using vectors to draw shapes



We can connect the tips of the vectors to create a shape, in this case a polygonal approximation of a circle. To do so, we want to create a "custom shape" by specifying its point. We do this by writing `begin_shape()` followed by a series of `vertex(...)` commands in a loop. We finally draw the shape by writing `end_shape(CLOSE)`. Simply writing `end_shape()` would not close the polygon.

````python
create_canvas(400, 400)
translate(center)
begin_shape()
n = 40
for i in range(n):
    angle = remap(i, 0, n, 0, TWO_PI)
    radius = width/3 
    pos = direction(angle)*radius
    vertex(pos)
end_shape()
show()
````

Here we create a circle by doing a `for` loop and re-mapping the index `i` to an angle. We then create a position along the circle by scaling the vector `direction(angle)` (having length one) by multiplying it with the desired radius. Try to replace the radius with a random function and to vary the number of points.

### Curve interpolation



Technically when creating a polygon, we are doing linear interpolation between
its points by connecting them with straight lines. Alike P5js and Processing,
Py5canvas allows you to interpolate the points of a polygon with curves, which
gives a smoother looking result. To do so, just replace `vertex` with
`curve_vertex`.



````python
create_canvas(400, 400)
translate(center)
begin_shape()
curve_tightness(0.5)
random_seed(121)
n = 20
for i in range(n):
    angle = remap(i, 0, n, 0, TWO_PI)
    radius = random(100, width/2) #width/3  # 
    pos = direction(angle)*radius
    curve_vertex(pos)
    #circle(pos, 5)
end_shape(CLOSE)
show()
````

The result is a curve widely used in computer graphics called a
[&ldquo;Catmull-Rom Spline&rdquo;](https://en.wikipedia.org/wiki/Centripetal_CatmullRom_spline). It was co-invented, among many other ideas in computer
graphics, by a computer scientist [Ed Catmull](https://en.wikipedia.org/wiki/Edwin_Catmull), who went on to found Pixar.

The `curve_tightness(amount)` command allows you to control how smooth the curve is, with values near zero being very smooth (zero does not work well) and a value of `1` being like a polygon. You can uncomment the `circle` function above to see the points that are being interpolated by the curve.



## Vectors and turtles
Another way to think of vectors is using the turtle analogy that we looked at last time. It turns out that py5canvas provides a "turtle object" that can be used more conveniently than the examples we created in the last session (You will need the latest version of Py5canvas for this to work!)

````python
from py5canvas import *
create_canvas(350, 350)
t = create_turtle(center)
t.rotate(20)
t.forward(150)
t.rotate(-90) # or t.left(90)
t.forward(50)
t.rotate(-90)
t.forward(100)
t.rotate(90)  # or t.right(90)
t.forward(50)
t.draw_turtle() # Draws the arrow head
show()
````

Each movement of the turtle can be thought of as a vector between going from its previous position to the next. Here `t.rotate(...)` sets the orientation of the vector and `t.forward(...)` sets its length. By default, the turtle draws the path of its motion. Each line segment visually represents a vector describing the turtle's motion. 



The turtle also has a "pen" that is touching the canvas by default. You can raise the pen with `t.pu()` (pen up) and lower it again with `t.pd()` (pen down), just like a virtual drawing machine. 



````python
create_canvas(400, 400)
t = create_turtle(center)
for i in range(200):
    t.rotate(random(-90, 90))
    t.forward(random(2, 10))
t.pu()
t.goto(100, 100)
t.pd()
t.angle(0)
t.square(50)
show()
````

Where `t.goto(x, y)` "teleports" the turtle to a given position,  `t.angle(ang)` sets the current angle and `t.square(size)` draws a square of a given size. 

A drawing like the one in the robot image can be created by using the `t.circle(radius, steps)` command, which approximates a circle with the desired number of segments:

````python
create_canvas(400, 400)
background(0)
stroke(255)
t = create_turtle(center)
n = 20
for i in range(n):
    t.right(360/n)
    t.circle(90, steps=7)
show()
````

## Vectors, sine and cosine

Recall, that it is useful to think of a vector simply as an arrow with a length and an orientation. To draw a circle in the example above, we incrementally took increasing angles and created a vector along that angle and scaled it by the radius. If we consider a circle with radius $=1$, the x and y coordinates of a point along the circle are given by the cosine and sine function respectively. As the angle (measured in radians) changes, cosine gives the x position and sine gives the y position. In Py5canvas these are called `cos` and `sin`. 

You might recall something similar to this from highschool



In this figure, our vector would have its tail on B and its tip on A, and the hypothenuse is the radius.


In practice, we could easily write the `direction(angle)` function as `vector(cos(angle), sin(angle))`, which gives us the vector (of length one) pointing along a circle with radius one. So multiplying this by the radius, gives us points along the circle with the desired radius. 

In the following example we define a custom shape using `begin_shape()` and `end_shape()`. Writing  

````python
create_canvas(400, 400)
translate(center)
begin_shape()
n = 40
for i in range(n):
    angle = remap(i, 0, n, 0, TWO_PI)
    radius = width/3 
    pos = vector(cos(angle), sin(angle))*radius
    vertex(pos)
end_shape(CLOSE)
show()
````

Specifying vectors (or coordinates) in this way, can be very powerful for generated graphics and is known as a polar coordinate system, where points are expressed by an angle and length, rather than the traditional (Cartesian) x and y coordinates. 



The expression `direction(angle)*length` gives us a 2D vector expressed in Cartesian coordinates given the 2D polar coordinates (angle, length). To retrieve the polar coordinates for a vector `v` you would do `heading(v)` for the angle and `mag(v)` for the length. 



>NOTE In a polar coordinate system, you cant directly add or subtract points the way you would with Cartesian coordinates. Angles and lengths dont combine linearly  for example, adding two polar coordinates doesnt give you a simple geometric sum. If you need to combine vectors, its usually best to convert them to their Cartesian components (x and y), perform the addition, and then convert back to polar if needed.

## Sine and cosine as functions
You can go around a circle as many time as you like in a periodic manner, and as this happens the cosine and sine functions produce an oscillatory behavior. We will look at this more in detail in the next class, but these functions become very useful to produce oscillatory patterns in your graphics. Let's visualize both:

````python
create_canvas(500, 300)
n = width//8 # <- note this is a division that always gives an integer
phase = 0 # PI/2 <- replace zero with this to align
stroke(0, 30)
for i in range(n):
    t = remap(i, 0, n, 0, TWO_PI*4) #<- increase it
    x = remap(i, 0, n, 0, width)
    
    line(x, 0, x, height)
    circle(x, cos(t)*50+70, 8)
    circle(x, sin(t + phase)*50+220, 8)
    
show()
````

You can see that sine and cosine are wave-forms where one is a shifted (phased) version of the other. They are shifted by exactly `PI/2` radians, which is 90 degrees. So think of a point going along the circle and keep track of the X and Y positions as it moves. The X (horizontal) position is the cosine and the Y (vertical) is the sine, and the rotation from from horizontal to vertical is 90 degrees.

## Sine and cosine as waves
We can view sine and cosine simply as functions that take an input number and output smooth oscillating values, which can become a powerful tool for generating graphics (or even audio).

The shift of a value inside the sine or cosine function, or any repeating waveform, is called the **phase**.  
With sine and cosine, this phase is expressed in **radians**.  
Adding or subtracting a value inside the parentheses shifts the wave horizontally.

For example we have seen that  `sin(x + PI/2)` is the same as `cos(x)`, a sine wave (phase) shifted by 90 degrees.



### Phase

The phase determines *where* the wave starts.  
Adding a positive value inside the function (`sin(x + phase)`) shifts the wave to the **left**,  
while subtracting (`sin(x - phase)`) shifts it to the **right**.



### Frequency

Multiplying the argument inside the function changes how *often* the wave repeats over a given interval.

- `sin(2 * x)` has **twice the frequency** of `sin(x)`, it completes two full cycles in the same space.  
- `sin(0.5 * x)` has **half the frequency**, it stretches out and repeats more slowly.

### Amplitude

Multiplying the *entire function* changes the **amplitude**, or the *height* of the wave.

- `2 * sin(x)` doubles the amplitude  the peaks and troughs are twice as tall.  
- `0.5 * sin(x)` reduces the amplitude by half.


Experiment with this code by changing the phase, frequency (`freq`) and amplitude (`amp`) variables, switching between `sin` and `cos`

````python
create_canvas(300, 200)
translate(0, height/2) # Vertical center because sin/cos are also negative
no_fill()
begin_shape()
n = 100
freq = 4 
amp = height/2
phase = 0.0
for i in range(100):
    t = remap(i, 0, n-1, 0, TWO_PI)
    x = remap(i, 0, n-1, 0, width)
    vertex(x, cos(t*freq + phase)*amp)
end_shape()
show()
````

# Back to vectors, the dot product 
Cosine and sine are related by the 90 degree (`PI/2` radians) rotation and vectors and cosines are related by the "dot product". 

Vectors and the cosine function are also connected through the dot product. The dot product is a special operation between two vectors: you multiply their corresponding components and then add those products together, for two vectors $(x_{1}, y_{1})$ and $(x_{2}, y_{2})$ this would be $x_1x_2 + y_1y_2$, or in code:
```
x1*x2 + y1*y2
```
You can use the function `dot(a, b)` to compute it for two vectors `a` and `b`.


Geometrically, it measures how much one vector is aligned with another, or how long is the projection of one vector is on the other. So for instance, if two vectors point in opposite directions, the dot product will be negative, and if the are perpendicular the dot product will be zero



Now it turns out that `dot(a,b)/(length(a)*length(b))` is the same as the cosine of the angle between `a`, and `b`. As a result, if the vectors are both of length one, then their dot product *is* the cosine of their angle. This elegant relation ties numbers to geometry in  


````python
create_canvas(300, 300)
a = vector(100, 20)
b = vector(-50, 30)
translate(center)
draw_vector(a)
draw_vector(b)
stroke(255, 0, 0)
d = dot(a, b)/mag(a)
draw_vector((a/mag(a))*d)
print(d)
show()
````
