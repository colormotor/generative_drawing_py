## Taking a line for a walk

A group expercise based on sketch by Simon Katan.

First execute the following cell to set this up first:

```python
from py5canvas import *

def next_cell():
    global cur_cell
    # This sets our position to the current cell and incements
    # cells are organized in a 4x4 grid
    reset_matrix()
    col = cur_cell%4
    row = (cur_cell - col)/4
    translate(cell_size*col, cell_size*row)
    cur_cell += 1

def init_grid():
    global cell_size, cur_cell
    cell_size = width/4
    push()
    no_fill()
    stroke_weight(0.5)
    stroke(0,100)
    cur_cell = 0
    # reset to cell 0
    for i in range(0, 16):
        next_cell()
        rect(0, 0, cell_size, cell_size)
    pop()
    cur_cell = 0
```

## Line tracing



- Consider [John Franzen's process](https://www.johnfranzen.com/each-line-one-breath)
- Read Andrew Hoff's [description](https://inconvergent.net/generative/linetrace/)
- Try to recreate line tracing from this template. 
- Try to make the code act in the same way as John Franzen draws his lines

```python
def draw_path(points):
    # Draws a curve connecting a list points [[x1, y1], [x2, y2], ...]
    begin_shape()
    for j in range(len(points)):
        curve_vertex(points[j][0], points[j][1])
    end_shape()

create_canvas(600, 500)

line_spacing = 5       # the distance that the vertical lines are apart
point_spacing = 10     # the distance between points on a line

no_fill()
background(255)
stroke(0)

num_points = int((height / point_spacing) * 0.75)

# initial pen position
x, y = 0, 0

# points for the first line
line_pts = []
for _ in range(num_points):
    line_pts.append([x, y])
    y += point_spacing

# draw the first line
draw_path(line_pts)

# draw the remaining lines
while x < width:
    # start at the top
    y = 0
    # move to the right
    x += line_spacing

    # create the positions for the new line
    line_pts = []
    for _ in range(num_points):
        line_pts.append([x, y])
        y += point_spacing

    draw_path(line_pts)

show()
```
